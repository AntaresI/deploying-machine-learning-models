```python
import plotly
import plotly.express as px
```

```python
def log_transform(df,var):
    df = df.copy()
    df[var] = np.log(df[var])
    fig = px.histogram(df, x=df[var],title=var)
    fig.show()
    return df
```

```python
for i in ['Balance', 'Out']:
    data = log_transform(data,i)
```

```python
data.columns
```

```python
subset_code = data.groupby('code')['Out'].median()
```

```python
subset_code
```

```python
def create_subset(*,df, code) -> pd.DataFrame:
    df = df.copy()
    df = df[df['code']==code]
    df = df[['date', 'Out']]
    return df
```

```python
def analyse_month(*,df,month)-> pd.DataFrame: 
    df = df.copy()
    df = df[df['date']==month]
    df = df['Out']
    return df
```

```python
for i in subset_code.index:  
    #if subset_code[i]<5 or subset_code[i] > 8:
     #   print('Skip')
      #  continue
    subset = create_subset(df=data,code=i)
    fig = px.box(subset, x='date',y='Out')
    fig.show()

    sub = analyse_month(df=subset,month='8')
    #print(sub)
    fig = px.histogram(sub, height=300,width=500,title=i)
    
    fig.show()
    
```
